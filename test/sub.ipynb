```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the file path
file_path = "/Users/Admin/Documents/Math/FoodAccessResearchAtlasData2019.csv"

# Load the CSV file into a DataFrame and display the first few rows
df = pd.read_csv(file_path)
df.head(147)



# File path to the dataset
file_path = "dataset/FoodAccessResearchAtlasData2019.csv"

def plot_median_income():
    """
    Plots a bar graph of the average Median Family Income per State using data from the specified file.
    """
    # Load the dataset
    dataframe = pd.read_csv(file_path)
    
    # Check if the necessary columns exist
    if 'State' not in dataframe.columns or 'MedianFamilyIncome' not in dataframe.columns:
        print("Error: Required columns not found in the dataframe.")
        return
    
    # Group by state and calculate the average median family income
    state_income = dataframe.groupby('State')['MedianFamilyIncome'].mean().reset_index()
    
    # Sort values for better visualization
    
    # Set the figure size
    plt.figure(figsize=(12, 6))
    
    # Create bar plot
    sns.barplot(x='State', y='MedianFamilyIncome', data=state_income, palette='viridis', hue=0)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)
    
    # Set labels and title
    plt.xlabel("State")
    plt.ylabel("Average Median Family Income")
    plt.title("Average Median Family Income by State")
    
    # Show the plot
    plt.show()

# Example usage:
plot_median_income()
```

```python
def plot_tract_low_income():
    """
    Plots a bar graph comparing TractLOWI by state.
    """
    # Load the dataset
    dataframe = pd.read_csv(file_path)
    
    # Check if the necessary column exists
    if 'State' not in dataframe.columns or 'TractLOWI' not in dataframe.columns:
        print("Error: Required columns not found in the dataframe.")
        return
    
    # Group by state and calculate the average TractLOWI
    state_data = dataframe.groupby('State')['TractLOWI'].mean().reset_index()
    
    # Set figure size
    plt.figure(figsize=(12, 6))
    
    # Create bar plot
    sns.barplot(x='State', y='TractLOWI', data=state_data, palette='viridis', legend=False, hue=0)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)
    
    # Set labels and title
    plt.xlabel("State")
    plt.ylabel("Average TractLOWI")
    plt.title("Average Low-Income Tracts by State")
    
    # Show the plot
    plt.show()

plot_tract_low_income()
```

```python
def plot_low_income_access():
    """
    Plots three bar graphs comparing LALOWI1_10, LALOWI05_10, and LALOWI1_20 by state.
    """
    # Load the dataset
    dataframe = pd.read_csv(file_path)
    
    # Check if the necessary columns exist
    required_columns = ['State', 'LALOWI1_10', 'LALOWI05_10', 'LALOWI1_20']
    if not all(col in dataframe.columns for col in required_columns):
        print("Error: Required columns not found in the dataframe.")
        return
    
    # Group by state and calculate the average values
    state_data = dataframe.groupby('State')[['LALOWI1_10', 'LALOWI05_10', 'LALOWI1_20']].mean().reset_index()
    
    # Set figure size
    fig, axes = plt.subplots(3, 1, figsize=(12, 18))
    
    # Define column names and titles
    columns = ['LALOWI1_10', 'LALOWI05_10', 'LALOWI1_20']
    titles = ["Low Income & Low Access 1 and 10 Mile", "Low Income & Low Access 1.5 and 10 Mile", "Low Income & Low Access 1 and 20 Mile"]
    
    # Create subplots
    for ax, col, title in zip(axes, columns, titles):
        sns.barplot(x='State', y=col, data=state_data, palette='viridis', ax=ax, hue=0)
        ax.tick_params(axis='x', rotation=90)
        ax.set_xlabel("State")
        ax.set_ylabel("Population")
        ax.set_title(title)
    
    # Adjust layout and show plot
    plt.tight_layout()
    plt.show()

plot_low_income_access()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the CSV files into DataFrames
df = pd.read_csv('dataset/ACCESS.csv')
supplement_df = pd.read_csv('dataset/Supplemental Data - State.csv')

# Group the data by 'State' and calculate the average for the specified columns
state_avg = df.groupby('State')[['PCH_LACCESS_POP_10_15', 'LACCESS_POP15', 'LACCESS_POP10']].mean().reset_index()

# Rename the columns for clarity
state_avg.columns = ['State', 'Avg PCH_LACCESS_POP_10_15', 'Avg LACCESS_POP15', 'Avg LACCESS_POP10']

# Merge the two DataFrames on the 'State' column
merged_df = pd.merge(state_avg, supplement_df, on='State', how='outer')

# Calculate the differences between the existing data and supplement data
for col in ['LACCESS_POP15', 'LACCESS_POP10']:
    if col in supplement_df.columns:
        merged_df[f'{col}_Diff'] = merged_df[f'Avg {col}'] - merged_df[col]
    else:
        print(f"Warning: '{col}' column not found in supplement data. Skipping this comparison.")

# Display the merged DataFrame with differences
print(merged_df[['State', 'Avg LACCESS_POP15', 'Avg LACCESS_POP10',
                 'LACCESS_POP15', 'LACCESS_POP10', 'LACCESS_POP15_Diff', 'LACCESS_POP10_Diff']])

# Calculate and display summary statistics of the differences
diff_columns = [col for col in merged_df.columns if col.endswith('_Diff')]
diff_summary = merged_df[diff_columns].describe()
print("\nSummary statistics of differences:")
print(diff_summary)

# Identify states with significant differences (e.g., more than 2 standard deviations)
significant_diff = merged_df[
    (np.abs(merged_df['LACCESS_POP15_Diff']) > 2 * diff_summary.loc['std', 'LACCESS_POP15_Diff']) |
    (np.abs(merged_df['LACCESS_POP10_Diff']) > 2 * diff_summary.loc['std', 'LACCESS_POP10_Diff'])
]

print("\nStates with significant differences:")
print(significant_diff[['State', 'LACCESS_POP15_Diff', 'LACCESS_POP10_Diff']])

# Visualize the differences
plt.figure(figsize=(12, 6))
sns.boxplot(data=merged_df[diff_columns])
plt.title('Distribution of Differences Between Calculated and Supplemental Data')
plt.ylabel('Difference')
plt.show()

# Scatter plot to compare calculated vs supplemental data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.scatter(merged_df['Avg LACCESS_POP15'], merged_df['LACCESS_POP15'])
ax1.set_xlabel('Calculated LACCESS_POP15')
ax1.set_ylabel('Supplemental LACCESS_POP15')
ax1.set_title('LACCESS_POP15: Calculated vs Supplemental')
ax1.plot([merged_df['LACCESS_POP15'].min(), merged_df['LACCESS_POP15'].max()], 
         [merged_df['LACCESS_POP15'].min(), merged_df['LACCESS_POP15'].max()], 
         'r--', lw=2)

ax2.scatter(merged_df['Avg LACCESS_POP10'], merged_df['LACCESS_POP10'])
ax2.set_xlabel('Calculated LACCESS_POP10')
ax2.set_ylabel('Supplemental LACCESS_POP10')
ax2.set_title('LACCESS_POP10: Calculated vs Supplemental')
ax2.plot([merged_df['LACCESS_POP10'].min(), merged_df['LACCESS_POP10'].max()], 
         [merged_df['LACCESS_POP10'].min(), merged_df['LACCESS_POP10'].max()], 
         'r--', lw=2)

plt.tight_layout()
plt.show()

# Calculate correlation coefficients
corr_15 = merged_df['Avg LACCESS_POP15'].corr(merged_df['LACCESS_POP15'])
corr_10 = merged_df['Avg LACCESS_POP10'].corr(merged_df['LACCESS_POP10'])

print(f"\nCorrelation coefficient for LACCESS_POP15: {corr_15:.4f}")
print(f"Correlation coefficient for LACCESS_POP10: {corr_10:.4f}")

# Calculate percentage difference
merged_df['LACCESS_POP15_Pct_Diff'] = (merged_df['LACCESS_POP15_Diff'] / merged_df['LACCESS_POP15']) * 100
merged_df['LACCESS_POP10_Pct_Diff'] = (merged_df['LACCESS_POP10_Diff'] / merged_df['LACCESS_POP10']) * 100

print("\nAverage percentage difference:")
print(f"LACCESS_POP15: {merged_df['LACCESS_POP15_Pct_Diff'].mean():.2f}%")
print(f"LACCESS_POP10: {merged_df['LACCESS_POP10_Pct_Diff'].mean():.2f}%")

# Identify states with the largest percentage differences
print("\nStates with the largest percentage differences:")
print(merged_df.nlargest(5, 'LACCESS_POP15_Pct_Diff')[['State', 'LACCESS_POP15_Pct_Diff']])
print(merged_df.nlargest(5, 'LACCESS_POP10_Pct_Diff')[['State', 'LACCESS_POP10_Pct_Diff']])
```
