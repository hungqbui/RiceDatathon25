```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, pointbiserialr, spearmanr, kendalltau
```

```python
fe_df = pd.read_csv("dataset/LOCAL.csv")
numeric_columns = fe_df.select_dtypes(include=['number']).columns


# Calculate the correlation matrix
correlation_matrix = fe_df[numeric_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(40, 32))

# Create a mask to hide the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create the heatmap
sns.heatmap(correlation_matrix, 
            mask=mask,
            annot=True, 
            cmap='coolwarm', 
            linewidths=0.5, 
            fmt='.2f',
            square=True,
            cbar_kws={"shrink": .8},
            annot_kws={'size': 8})

# Set the title
plt.title('Correlation Heatmap of Numeric Features', fontsize=16, pad=20)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
